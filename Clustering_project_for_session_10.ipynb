```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate synthetic data
np.random.seed(0)
n_samples = 200
n_features = 2

# Create two clusters of scenarios
cluster1 = np.random.randn(n_samples // 2, n_features) + [2, 2]
cluster2 = np.random.randn(n_samples // 2, n_features) + [-2, -2]
data = np.vstack((cluster1, cluster2))

# Create a DataFrame
df = pd.DataFrame(data, columns=['Threat Level', 'Environmental Hazard'])

# Perform K-means clustering
kmeans = KMeans(n_clusters=2, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[['Threat Level', 'Environmental Hazard']])

# Visualize the clusters
plt.figure(figsize=(8, 6))
colors = ['blue', 'red']
for cluster, color in zip(range(2), colors):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(
        cluster_data['Threat Level'],
        cluster_data['Environmental Hazard'],
        c=color,
        label=f'Cluster {cluster}',
    )

plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    s=300,
    c='yellow',
    marker='*',
    label='Cluster Centers',
)
plt.xlabel('Threat Level')
plt.ylabel('Environmental Hazard')
plt.legend()
plt.title('Scenario Clustering')
plt.show()

# Define suit adaptation for each cluster
cluster_actions = {
    0: "Defensive Mode",
    1: "Offensive Mode",
}

# Apply the suit adaptation logic based on the cluster
df['Suit_Mode'] = df['Cluster'].map(cluster_actions)

# Display the DataFrame with cluster labels and suit modes
print(df)
```

```python
import pandas as pd
import numpy as np

# Generate synthetic data for Threat Level and Environmental Hazard
np.random.seed(0)
n_samples = 200
n_features = 2

# Create two clusters of scenarios
cluster1 = np.random.randn(n_samples // 2, n_features) + [2, 2]
cluster2 = np.random.randn(n_samples // 2, n_features) + [-2, -2]
data = np.vstack((cluster1, cluster2))

# Create a DataFrame
df = pd.DataFrame(data, columns=['Threat Level', 'Environmental Hazard'])

# Perform K-means clustering
kmeans = KMeans(n_clusters=2, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[['Threat Level', 'Environmental Hazard']])

# Define suit adaptation for each cluster
cluster_actions = {
    0: "Defensive Mode",
    1: "Offensive Mode",
}

# Apply the suit adaptation logic based on the cluster
df['Suit_Mode'] = df['Cluster'].map(cluster_actions)

# Add more rows to achieve a total of 200 rows
extra_rows = 200 - df.shape[0]
if extra_rows > 0:
    additional_data = np.random.randn(extra_rows, n_features)
    additional_df = pd.DataFrame(additional_data, columns=['Threat Level', 'Environmental Hazard'])
    additional_df['Cluster'] = kmeans.predict(additional_data)
    additional_df['Suit_Mode'] = additional_df['Cluster'].map(cluster_actions)
    df = pd.concat([df, additional_df], ignore_index=True)

# Save the DataFrame to a CSV file
df.to_csv('iron_man_suit_data.csv', index=False)

# Print the total number of rows
print("Total Rows:", df.shape[0])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

# Set the number of rows and columns
num_rows = 300
min_num_columns = 5
max_num_columns = 100

# Define the column names
column_names = [
    'ID', 'Armor_Class', 'Power_Source', 'Weaponry', 'Speed',
    'Strength', 'Threat_Level', 'Environmental_Hazard', 'Suit_Mode'
]

# Generate random data for the dataset
data = {
    'ID': np.arange(1, num_rows + 1),
    'Armor_Class': np.random.choice(['Mark I', 'Mark II', 'Mark III', 'Mark IV'], num_rows),
    'Power_Source': np.random.choice(['Arc Reactor', 'Fusion Reactor', 'Nuclear Reactor'], num_rows),
    'Weaponry': np.random.choice(['Repulsor Beams', 'Unibeam', 'Missiles', 'Lasers'], num_rows),
    'Speed': np.random.choice(['Slow', 'Moderate', 'High'], num_rows),
    'Strength': np.random.choice(['Low', 'Moderate', 'High'], num_rows),
    'Threat_Level': np.random.randint(1, 5, num_rows),
    'Environmental_Hazard': np.random.choice(['Low', 'Medium', 'High'], num_rows),
    'Suit_Mode': np.random.choice(['Defensive Mode', 'Offensive Mode'], num_rows)
}

# Create the DataFrame
df = pd.DataFrame(data, columns=column_names)

# Perform label encoding for 'Environmental_Hazard'
label_encoder = LabelEncoder()
df['Environmental_Hazard_Label'] = label_encoder.fit_transform(df['Environmental_Hazard'])

# Perform K-means clustering
# Select relevant columns for clustering
clustering_columns = ['Threat_Level', 'Environmental_Hazard_Label']
kmeans = KMeans(n_clusters=2, random_state=0)
df['Cluster'] = kmeans.fit_predict(df[clustering_columns])

# Visualize the clusters
plt.figure(figsize=(8, 6))
colors = ['blue', 'red']
for cluster, color in zip(range(2), colors):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(
        cluster_data['Threat_Level'],
        cluster_data['Environmental_Hazard_Label'],
        c=color,
        label=f'Cluster {cluster}',
    )

plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    s=300,
    c='yellow',
    marker='*',
    label='Cluster Centers',
)
plt.xlabel('Threat Level')
plt.ylabel('Environmental Hazard (Encoded)')
plt.legend()
plt.title('Clustering of Iron Man Suits')
plt.show()

# Save the dataset to a CSV file with cluster labels
df.to_csv('ironman_suit_db_with_clusters.csv', index=False)
```
